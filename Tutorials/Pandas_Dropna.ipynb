```python
import pandas as pd
```

```python
csv_data = pd.read_csv('drop.csv')
```

```python
csv_data
```

#### drop all rows that have any NaN values

```python
csv_data.dropna() # default values => axis=0, how='any'
```

Only row 0 has no NaN value that why only that row left

#### drop only if ALL columns are NaN

```python
csv_data.dropna(how='all') # means drop row if its all column are NaN
```

In row 1 there is all column data is NaN so it droped

#### Drop only if all row are NaN

```python
csv_data.dropna(how='all', axis=1) # means drop column if its all row values are NaN
```

There is no such column which has alll row are NaN

#### Drop row if it does not have at least "n" values that are **not** NaN in row

```python
csv_data.dropna(thresh=10) #means atleast 10 valuse in a row should not be NaN
```

In above table row 1 and 5 does not fullfil the criteria of 10 values not NaN

#### Drop row if it does not have at least "n" values that are **not** NaN in column

```python
csv_data.dropna(axis=1, thresh=4) #means atleast 4 valuse in a column should not be NaN
```

You can see Column Embarked droped since it does not fulful criteria

#### Drop only if NaN in specific column

```python
csv_data.dropna(subset=['Parch']) #means if  any row in Parch column is NaN then delete that row
```

In Parch column row 1 & 5 has NaN value so they droped

#### Drop only if NaN in specific row

```python
csv_data.dropna(axis=1,subset=[5]) #means if  any column in row 5 is NaN then delete that column
```

As you can see row 5 has NaN values in column ['Name', 'Sex', 'Parch'] so these column droped
